### მონაცემების ერთ ხაზზე შეტანა

აქამდე მონაცემები (`num1, num2`) შეგვყავდა ახალ-ახალი ხაზიდან (ცალ-ცალკე), ჯერ პირველი რიცხვი, შემდეგ `enter`-ის დაჭერის შემდეგ -- მეორე. შესაძლებელია მონაცემების შეტანა ერთ ხაზზე.

დავუშვათ, გვინდა ორი რიცხვი შევიტანოთ ერთი ხაზიდან: ვკრებთ ჯერ პირველ რიცხვს, შემდეგ -- ინტერვალს, ვაგრძელებთ მეორე რიცხვის შეტანას და ბოლოს ვაფიქსირებთ `enter`-ით. ამ შემთხვევაში `num1` ცვლადში მოთავსდება პირველი რიცხვი, `num2`-ში კი -- მეორე. ჩავწეროთ პროგრამის ფრაგმენტი:

````
print("შეიტანეთ ორი მთელი რიცხვი:")
s = input()
num1, num2 = s.split()

num1 = int(num1)
num2 = int(num2)

print("შტანილი რიცხვები:", num1, num2)
````

`input()` ფუნქციის საშუალებით `s` ცვლადში მოთავსდება ეკრანიდან აკრეფილი ორი რიცხვი და მათ შორის ინტერვალი. მიაქციეთ ყურადღება, რომ ცვლადში ინახება ტექსტი: პირველი რიცხვი, ინტერვალი (`space`) და მეორე რიცხვი. `int(s)` ჩაწერა არ შეიძლება, რადგან `s` ცლადში წერია ტექსტი და ორი რიცხვი.

ამ რიცხვების განსაცალკავცებლად გამოიყენება ფუნქცია `split()`. თუ ფრჩხილებში არ ჩავწერთ არაფერს, იგულისხმება რომ გამყოფი სიმბოლო არის ან `space`, ან `enter <===> '\n'` ან ტაბულაცია `\t` (ოთხი `space`). ჩვნენ შემთხვევაში რიცხვების გამყოფი არის ინტერვალი.

`num1, num2 = s.split()` ნიშნავს რომ აკრეფილი რიცხვები დაიყოფა ორ ნაწილად და ცვლადებში მოთავსდება ორივე რიცხვი შესაბამისად `num1`-ში პირველი და `num2`-ში მეორე. შემდეგ ტექსტურ რიცხვებს გარდავქმნით მთელ რიცხვებად (`num1 = int(num1) და  num2 = int(num2)`) და ვბეჭდავთ.

`num1, num2 = s.split()` გამოსახულებაში გამოიყენება *„მრავალმხრივი მინიჭება“* -- ერთ გამოსახულებაში ხდება ორი ცვლადისთვის მნიშვნელობების მინიჭება.

შესაძლებელია პროგრამის ფრაგმენტი უფრო კომპაქტურად ჩავწეროთ:

````
print("შეიტანეთ ორი მთელი რიცხვი:")
num1, num2 = input().split()

num1 = int(num1)
num2 = int(num2)

print("შტანილი რიცხვები:", num1, num2)
````

ან კიდევ უფრო მოკლედ:

````
print("შეიტანეთ ორი მთელი რიცხვი:")
num1, num2 = map(int, input().split())

print("შტანილი რიცხვები:", num1, num2)
````

აქ გამოიყენება ფუნქცია `map()`, რომელიც არგუმენტად იყენებს სხვა ფუნქციას, ჩვენ შემთხვევაში `int`, მეორე არგუმენტად კი ეკრანიდან შეტანილ ინფორმაციას, რომელიც დაყოფილია ორ ნაწილად `input().split()`. ეს დაყოფილი რიცხვები, როგორც მთელი რიცხვები, მიენიჭება ცვლადებს.

*„მრავალმხრივი მინიჭებისას“* უნდა გავითვალისწინოთ ის, რომ მინიჭების მარცხენა მხარეს ჩაწერილი ცვლადების რაოდენობა უნდა უმთხვეოდეს ეკრანიდან (კლავიატურიადან) აკრეფილ მონაცემთა რაოდენობას `space`-ების გამოკლებით. ვკრებთ ორი რიცხვს, ე.ი. მინიჭების მარცნივ ორი ცვლადი უნდა გვქონდეს: არც მეტი, არც ნაკლები. წინააღმდეგ შემთხვევაში აღიძვრება შეცდომა.

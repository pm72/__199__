### არითმეტიკული გამოსახულება

* მიმატება +  
* გამოკლება -  
* გამრავლება *  
* გაყოფა /
* მთელზე გაყოფა //  
* ნაშთის აღება %  
* ახარისხება **  

თავიდან განვიხილოთ ოპერაციები ერთ ცვლადზე. მინიჭების ოპერაცია:  
`i = i + 1`  
`i` ცვლადს უმატებს `1`-ს. ეს ოპერაცია ხშირად გამოიყენება. მაგ.: ციკლისას.

მინიჭება უფრო რთული გამოსახულებისას:

\begin{equation}
a = \frac{c + b - 1}{2} \cdot d
\end{equation}

ჩაიწერება ასე:  
`a = (c + b - 1) / 2 * d`

ასეთ ჩანაწერს არითმეტიკული გამოსახულების _წრფივ ჩანაწერს_ უწოდებენ (ერთ ხაზზე ჩაწერა). გრძელი ფორმულის ჩაწერისას შესაძლებელი ფორმულა გავყოთ მაგალითად ორ ნაწილად (ორ ხაზზე ჩავწეროთ):

```python
a = x * (2*c + 4*b - 8*f) + \
2 * a * c
```

ფრჩხილების შიგნით გადატანისას `\` სიმბოლო საჭორო არ არის:

```python
a = (c +
 5 +
  2 *
   c)

a = (c + 5
+ 2 * c)
```

გამოსახულებაში არითმეტიკული ოპერაციების მოქმედება ხორციელდება _პრიორიტეტული_ ოპერაციით:
* ფრჩხილებში მოთავსებული ოპერაციები სრულდება პირველად
* შემდეგი პრიორიტეტი აქვს ახარისხებას, მარჯვნიდან მარცხნივ: `2**3**2 <===> 2**(3**2)`
* შემდეგია გამრავლება და გაყოფა, მარცხნიდან მარჯვნივ
* ბოლოს შეკრება-გოკლება, მარცხნიდან მარჯვნივ

ამრიგად, გამრავლებას და გაყოფას აქვს ერთნაირი პრიორიტეტი, უფრო მაღალი, ვიდრე შეკრება-გამოკლებას.

მაგალითად, გამოსახულებაში  
`a = (c + b - 1) / 2 * d`  
პირველად შესრულდება ფრჩხილებში მოთავსებული ოპერაცია: მიმატება და შემდეგ გამოკლება; შემდეგ გაყოფა 2-ზე და ბოლოს -- გამრავლება `d`-ზე.

`Python`-ში შესალებელია _მრავალმხრივი_ მინიჭება:  
    `a, b = 1, 2 `
იგივეა, რაც  
    `a = 1`  
    `b = 2`  
ითვლება, რომ ეს ორი მოქმედება ხორციელდება პარალელურად, ერთდროულად.

გამოსახულება `a = b = 0` ნიშნავს:
```python
b = 0
a = 0
```

ცვლადის მნიშვნელობის შესაცვლელად, მოსახერხებელია _შემოკლებული_ ჩანაწერის გამოყენება:

```python

a = a + b             a += b
a = a - b             a -= b
a = a * b             a *= b

```

### მთელზე გაყოფა

გაყოფისას მიღებული შედეგი არ არის მთელი, ასეთ რიცხვებს უწოდებენ _ნამდვილი რიცხვები_, ათწილადი რიცხვები. პროგრამირებაში მათ _მცოცავწერტილიანი რიცხვები_ ჰქვია (`floating point`).

ხშირად საჭიროა გაყოფისას მივიღოთ მთელი რიცხვი, ან ნაშთი. მაგალითად, ცნობილია დროის ინტერვალი წამებში, ვთქვათ 289 წამი, და გვინდა განვსაზღვროთ რამდენი წუთისა და წამისგან შედგება 289 წამი (289 წამი = 4 წთ. 49 წმ.). აქ 4 არის მთელი რიცხვი 289:60, 49 კი -- ნაშთი ამ გაყოფიდან.

ასეთ შემთხვევაში გამოიყენება სპეციალური ოპერატორები:

// -- მთელზე გაყოფა (ნაშთის უგულვებელყოფა)

% -- ნაშთის აღება გაყოფისას

```python
time_sec = 289
minutes = time_sec // 60    # = 4
seconds = time_sec % 60     # = 49
```

ამ ოპერაციების გამოყენებით მოსახერხებელია მუშაობა ცალკეულ ციფრებზე. ნებისმიერი მთელი რიცხვის 10-ზე გატოფისას მიღებული ნათში წარმოადგენს რიცხვის ბოლო ციფრს.
```python
N = 123
last_digit = N % 10    # = 3
```

თუ რიცხვს გავყობთ 10-ზე და ავიღებთ მხოლოდ მთელ ნაწილს, ჩვენ _„გადავაგდებთ“_ ბოლო ციფრს:
```python
N = 123
digits2 = N // 10    # = 12
```

საინტერესო შედეგს გვაძლევს `'//'` და `'%'` ოპერაცია უარყოფითი რიცხვების დროს:
```python
print(-7 // 2)     # = -4
print(-7 // 2)     # = 1
```

საქმე იმაშია, რომ ნაშთი ყოველთვის დადებითია. გაყოფისას ჩვენ ვეძებთ -7-ზე ყველაზე ახლოს მყოფ ნაკლებ უარყოფით რიცხვს, რომელიც უნაშთოთ იყოფა 2-ზე, ასეთი რიცხვის -8. ამიტომ `-7 // 2 = -4` და ნაშთი კი -- `-7 = -8 + 1`.

`python` ენაზე (განსხვავებით სხვა ენებისგან `C++, pascal, java ...`) ეს ოპერაციები მათემატიკურად სწორად სრულდება.

````
numb = 842     #   482
print(numb)
print(numb % 10, (numb // 10) % 10, numb // 100, sep='')

# print(12 + 5)
# print(12 - 5)
# print(12 * 5)
# print(12 / 5)     # ჩვეულებრივი, მათემატიკური გაყოფა
# print(12 // 5)    # მთელზე გაყოფა  => რამდენჯერ თავსდება რიცხვი (5)
# print(12 % 5)     # ნაშთი
````

### მონაცემების ეკრანზე გამოტანა

თქვენ უკვე იცით, რომ ფუნქცია `print()` გუმენტებს შორის _ინტერვალებს_ ბეჭდავს:
```python
a = 12
```

მაგრამ ჩვენ გვინდა ეს მონაცემები მოწესრიგებულად გამოვიტანოთ, მაგ.: ცხრილის სახით:
```python
  12    5 155
  211 315   8   
```
დავუშვათ, ვმუშაობთ ნატურალურ რიცხვებთან, რომლებიც 1000-ზე ნაკლებია. ყოველი რიცხვისთვის უნდა გამოვყოთ, ვთქვათ, ოთხი პოზიცია ეკრანზე: სამი რიცხვის ჩასაწერად და ერთი ინტერვალისთვის, რათა განვასვაოთ რიცხვები ერთმანეთისგან:
```python
print(f"{a:4}{b:4}{c:4}")
```
ასეთ ჩანაწერს _ფორმატირებული გამოტანა_ ეწოდება.

````
f_name = "პაატა"
l_name = "მამფორია"
age = 48

print(f_name, l_name, age, "წლის.")

s1 = f"{f_name} {l_name} {age} წლის."
print(s1)
````

სკოლაში გაკვეთილები იწყება 8:30. გაკვეთილი მიმდინარეობს 45 წუთი. მათ შორის დასვენება 5 წუთია. დაწერეთ პროგრამა, რომელიც მოგთხოვთ გაკვეთილის ნომერს და გამოიტანს მისი დამთავრების დროს ფორმატით: საათი:წუთი.

````
start = 8 * 3600 + 30 * 60
lesson = 45 * 60
antract = 5 * 60

n = 2

finish = start + lesson * n + antract * (n - 1)

print(f"{n} გაკვეთილის დამთავრების დრო: {finish // 3600:02d}:{finish // 60 % 60:02d}")
````

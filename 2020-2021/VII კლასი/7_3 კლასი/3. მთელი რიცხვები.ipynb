მინიჭება უფრო რთული გამოსახულებისას:

\begin{equation}
a = \frac{c + b - 1}{2} \cdot d
\end{equation}

ჩაიწერება ასე:  
`a = (c + b - 1) / 2 * d`

ასეთ ჩანაწერს არითმეტიკული გამოსახულების _წრფივ ჩანაწერს_ უწოდებენ (ერთ ხაზზე ჩაწერა). გრძელი ფორმულის ჩაწერისას შესაძლებელი ფორმულა გავყოთ მაგალითად ორ ნაწილად (ორ ხაზზე ჩავწეროთ):

```python
a = x * (2*c + 4*b - 8*f) \
+ 2 * a * c
```

ფრჩხილების შიგნით გადატანისას `\` სიმბოლო საჭორო არ არის:

```python
a = c + 5 + 2 * c
a = (c + 5 + 2 * c)

a = (c + 5
+ 2 * c)
```

გამოსახულებაში არითმეტიკული ოპერაციების მოქმედება ხორციელდება _პრიორიტეტული_ ოპერაციით:
* ფრჩხილებში მოთავსებული ოპერაციები სრულდება პირველად
* შემდეგი პრიორიტეტი აქვს ახარისხებას, მარჯვნიდან მარცხნივ: `2**3**2 <===> 2**(3**2)`
* შემდეგია გამრავლება და გაყოფა, მარცხნიდან მარჯვნივ
* ბოლოს შეკრება-გოკლება, მარცხნიდან მარჯვნივ

ამრიგად, გამრავლებას და გაყოფას აქვს ერთნაირი პრიორიტეტი, უფრო მაღალი, ვიდრე შეკრება-გამოკლებას.

`a = (c + b - 1) / 2 * d`

    `a, b = 1, 2 `

გაყოფისას მიღებული შედეგი არ არის მთელი, ასეთ რიცხვებს უწოდებენ _ნამდვილი რიცხვები_, ათწილადი რიცხვები. პროგრამირებაში მათ _მცოცავწერტილიანი რიცხვები_ ჰქვია (`floating point`).

საინტერესო შედეგს გვაძლევს `'//'` და `'%'` ოპერაცია უარყოფითი რიცხვების დროს:
```python
print(-7 // 2)     # = -4

```

საქმე იმაშია, რომ ნაშთი ყოველთვის დადებითია. გაყოფისას ჩვენ ვეძებთ -7-ზე ყველაზე ახლოს მყოფ ნაკლებ უარყოფით რიცხვს, რომელიც უნაშთოთ იყოფა 2-ზე, ასეთი რიცხვის -8. ამიტომ `-7 // 2 = -4` და ნაშთი კი -- `-7 = -8 + 1`.

19. დაწერეთ პროგრამა, რომელიც შეცვლის ავტომატის მუშაობას. ავტომატი შეტანისას ითხოვს ოთხნიშნა ნატურალურ რიცხვს. შემდეგ ითვლის სამ ოპერაციას: პირველი ორი ციფრის ჯამს, შუა ციფრების ჯამს და ბოლო ორი ციფრის ჯამს. გამოტანისას კი ბეჭდავს ოპერაციების ჯამს.

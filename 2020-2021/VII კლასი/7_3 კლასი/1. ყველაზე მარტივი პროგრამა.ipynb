`'#'` სიმბოლოთო დაწყებული სტრიქონი განიხილება, როგორც კომენტარი და ინტერპრეტატორის მიერ არ განიხილება შემსრულებელ საბრძანებო ინსტრუქციად. იგი მხოლოდ ინფორმაციის ხასიათს ატარებს. 

````
# ცარიელი პროგრამა
````

ახლა ჩვენს პროგრამას ვასწავლოთ გააკეთოს სასარგებლო რამ. მაგალითად დავაბეჭდინოთ ეკრანზე ტექსტი.  `Hello world!..`

````
print("Hello, world!..")
# print("დღეს უყურეთ ფილმს 'არმაგედონი'")
# print('დღეს უყურეთ ფილმს "არმაგედონი"')
# print('დღეს უყურეთ ფილმს "არმაგედონი"')
````

შესაძლებელია `print` ასეც ჩავწეროთ:

````
print("გამარჯობა", "ნიკა!")
````

`print`-ში სიტყვებს შორის ინტერვალი (`space`) ავტომატურად თავსდება.

თუ ის არ გვჭირდება, მაშინ `print` ფუნქციის გამოძახებისას, ბოლოში უნდა დავამატოთ კიდევ ერთი არგუმნეტი `sep`:

````
print("2", "+", "2", "=", "4", sep='')
````

ახლა შევეცადოთ რამდენიმე სტრიქონინანი ტექსტი გამოვიტანოთ ეკრანზე:

````
print("გამარჯობა მაკა!")
print("გამარჯობა ლიკა!")
````

როგორც ვხედავთ `print` ფუნქცია ახალი ხაზიდან ბეჭდავს ინფორმაციას (იყენებს `Enter`-ს).

თუ გვინდა რომ რამოდენიმე `print`-ის გამოყენებით ინფორმაცია არ იყოს ახალი ხაზიდან დაბეჭდილი, უნდა გამოვიყენოთ არგუმნეტი `end`:

````
print("1", end="")
print("23", end="")
print("456")
````

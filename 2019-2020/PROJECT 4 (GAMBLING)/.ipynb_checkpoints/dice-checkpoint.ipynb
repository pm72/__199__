```python
import random, time

print("გავაგოროთ კამათელი\n")

your_number = 0
my_number = 0
attempt = 0

while attempt < 5:
    attempt += 1
    print("რაუნდი {}".format(attempt))
    
    print("\tშენი გაგორებაა: ", end='')
    your_dice = random.randint(2, 12)
    time.sleep(1)
    print(your_dice)
    
    print("\tჩემი გაგორებაა: ", end='')
    my_dice = random.randint(2, 12)
    time.sleep(1)
    print(my_dice)
    
    if your_dice > my_dice:
        your_number += 1
    elif my_dice > your_dice:
        my_number += 1
    
    time.sleep(1)
    print("{} : {}".format(your_number, my_number))
    
    print()

    time.sleep(1)
if your_number > my_number: print("შენ მოიგე!")
elif my_number > your_number: print("მე მოვიგე!")
else: print("ყაიმი!")
```

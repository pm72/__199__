```python
alphabet = '0123456789ABCDEF'
base = len(alphabet)

for i in range(257):
    temp = i
    num = ''
    while temp != 0:
        num = alphabet[temp % base] + num
        temp //= base
    
    print("{}  <===>  {}".format(i, num))

    
# 5287  <===>  14A7 
# 5287 % 16
# 5287 // 16
# 330 % 16
# 330 // 16
# 20 % 16
# 20 // 16
# 1 % 16
# 1 // 16
```

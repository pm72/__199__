```python
from datetime import datetime

posts =[
    {
        'title': 'ჩემი პირველი სტატია',
        'content': 'ძაან მაგარი და საინტერესო სტატიაა',
        'author': 'hermano72',
        'date_pub': str(datetime.today().year) + "/" + str(datetime.today().month) + "/" + str(datetime.today().day) 
    }
] * 10

print('*' * 100)
for post in posts:
    print(post['title'])
    print(post['content'])
    print(post['author'])
    print(post['date_pub'], "\ n")
    print('*' * 100)
```
